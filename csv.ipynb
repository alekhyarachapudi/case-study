```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
aa=pd.read_csv('energy.csv')
print(aa)

srh=aa[aa['State']=='Andhra']
rcb=aa[aa['State']=='Karnataka']

fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'])
diag[1].plot(rcb['Date'],rcb['Consumption'])

srh=aa[aa['State']=='Telengana']
rcb=aa[aa['State']=='Tripura']

fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'])
diag[1].bar(rcb['Date'],rcb['Consumption'])

cse4='Kerala'
cse5='Delhi'
tdp=[cse4,cse5]

srh=aa[aa['State']=='Kerala']['Consumption'].sum()
rcb=aa[aa['State']=='Delhi']['Consumption'].sum()
ysr=[srh,rcb]

fig,diag=plt.subplots(2)
diag[0].pie(ysr,labels=tdp)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as cse

ntr=pd.read_csv('heart.csv')
print(ntr)

cse.scatterplot(x='age',y='bp',data=ntr,hue='gender')
plt.title('heart attack')
plt.xlabel('age')
plt.ylabel('bp')
plt.show()

cse.lineplot(x='age',y='cholestrol',data=ntr)
plt.title('line plot')
plt.show()

cse.boxplot(x='target',y='value',data=ntr)
plt.title('boxplot')
plt.show()

color=['magenta','pink']
cse.countplot(x='target',data=ntr,palette=color)
plt.show()

cse.pairplot(ntr)

```
